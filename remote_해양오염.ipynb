```python
import matplotlib.pyplot as plt
```

```python
X= [2016, 2017, 2018, 2019, 2020, 2021]
Y1 = [2654, 2980, 3230, 4020, 4411, 4682]
Y2 = [4574.2, 3835.4, 2585.1, 1698.6, 1014.9, 2899.4]
plt.plot(X, Y1, color='red', marker='o', alpha=0.5, linewidth=2)
plt.plot(X, Y2, color='blue', marker='o', alpha=0.5, linewidth=2)
 
#plt.title("플라스틱 사용량에 비례한 해안 쓰레기(kg)")
plt.xlabel("year")
plt.ylabel("red = t, blue = kg")
plt.show()
```
