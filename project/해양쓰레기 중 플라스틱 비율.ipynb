```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
X= [2016, 2017, 2018, 2019, 2020, 2021, 2022]
Y1 = [56.5, 58.1, 81.6, 80.2, 83.4, 88.5, 91]
plt.plot(X, Y1, color='red', marker='o', alpha=0.5, linewidth=2)
 
plt.title("국내 해안쓰레기 수량 중 플라스틱 비율")
plt.xlabel("연도")
plt.ylabel("플라스틱 수량 비율(%)")
plt.show()
```

```python
X= [2016, 2017, 2018, 2019, 2020, 2021, 2022]
Y1 = [38.6, 39.2, 58.8, 62.9, 46, 51.3, 49.2]
plt.plot(X, Y1, color='blue', marker='o', alpha=0.5, linewidth=2)
 
plt.title("국내 해안쓰레기 무게 중 플라스틱 비율")
plt.xlabel("연도")
plt.ylabel("플라스틱 무게 비율(%)")
plt.show()
```
