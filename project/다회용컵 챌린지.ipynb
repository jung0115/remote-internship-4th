```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
X= [1, 2, 3, 4, 5, 6]
Y1 = [4100, 3450, 2800, 2300, 1800, 1400]
Y2 = [2059, 1831, 1420, 2452, 1264, 525]
plt.plot(X, Y1, color='red', marker='o', alpha=0.5, linewidth=2)
plt.plot(X, Y2, color='blue', marker='o', alpha=0.5, linewidth=2)
 
plt.title("챌린지 참여도 비교")
plt.xlabel("개월")
plt.ylabel("red: 블로그 참여자수(백 명) / blue: 용기내 해시태그 수(개)")
plt.show()
```
